# **Data Visualization**

# **a) Univariate Visualization**

**Univariate Visualization** is analyzing and visualizing **only one variable   (column) at a time**. It helps us understand the distribution, spread, central tendency, and outliers of that variable.

# **1.Histogram for Profit Distribution**

We can start by visualizing the **Profit Distribution** using a **Histogram**. A histogram of profit shows how frequently different profit values occur across all sales transaction.

We can also overlay a **Kernel Density Estimate (KDE)** to smooth the
histogram and get a clearer sense of the data’s distribution.

```python
plt.figure(figsize=(6,4))
sns.histplot(df['profit'], bins=30, kde=True, color='green')
plt.title('Histogram for Profit Distribution')
plt.xlabel('profit')
plt.ylabel('Frequency')
plt.show()
```

# **2. Box Plot for Profit by Item Type**

We can visualize the **Profit distribution by Item Type** using a **Box Plot**.

Box plots show the median, quartiles, and potential outliers for each region.
This will help us compare the profit distribution across different item type and
identify any variations or anomalies.

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='item_type', y='profit', data=df, color='#f8cc1b')
plt.title("Boxplot for profit by Item Type")
plt.xticks(rotation=45)
plt.show()
```

# **Replace Outliers Based on IQR Method**

```python
# Similar to boxplot rules (Q1 - 1.5*IQR, Q3 + 1.5*IQR)

Q1 = df['profit'].quantile(0.25)
Q3 = df['profit'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df['profit'] = np.where(df['profit'] < lower_bound, lower_bound, df['profit'])
df['profit'] = np.where(df['profit'] > upper_bound, upper_bound, df['profit'])
```

# **Box Plot for Profit by Item Type**

```python
# After replace outliers
plt.figure(figsize=(8,5))
sns.boxplot(x='item_type', y='profit', data=df, color='#f37324')
plt.title("Boxplot for profit by Item Type")
plt.xticks(rotation=45)
plt.show()
```

# **3. Pie Chart for Transaction Type Distribution**
A **Pie Chart** breaks down total transactions based on payment method(transaction type). Each slice of the pie represents the percentage share of that payment mode.


```python
df['transaction_type'].value_counts().plot.pie(autopct='%1.1f%%', figsize=(6,6), colors=sns.color_palette("Set2"))
plt.title("Transaction Type Distribution")
plt.show()
```

# **13. Skewness Analysis for Profit**
**Skewness** is a measure of the asymmetry of a data distribution around its mean. It can be positive, negative, zero (including symmetry), or undefined. A **positive skew (right-skewed)** indicates a longer tail on the right side, while a **negative skew (left-skewed)** suggests a longer tail on the left side.

To analyze the skewness for the Profit column, we calculated the skewness value using the following code:

```python
# Check skewness for Profit
skewness = df[['profit']].skew()

# Display the skewness value
print(skewness)
```

```python
# Calculate mean, median, and mode for profit column
mean_profit = df['profit'].mean()
median_profit = df['profit'].median()
mode_profit = df['profit'].mode()[0]

# Plot histograms for Profit to visualize skewness
plt.figure(figsize=(8, 5))
sns.histplot(df['profit'], bins=30, kde=True, color='#660e60')
plt.axvline(mean_profit, color='blue', linestyle='dashed', linewidth=2, label=f'Mean: {mean_profit:.2f}')
plt.axvline(median_profit, color='red', linestyle='dashed', linewidth=2, label=f'Median: {median_profit:.2f}')
plt.axvline(mode_profit, color='yellow', linestyle='dashed', linewidth=2, label=f'Mode: {mode_profit:.2f}')
plt.title('Profit Distribution')
plt.legend()

# Fix layout issues
plt.tight_layout()
plt.show()
```

After calculating the skewness value for the **Profit** column, we obtain the following insights into the distribution:

**Profit Distribution**: The Profit Distribution is **positively skewed (right-skewed)**, with the mean (54.13) higher than the median (48.00) and mode (60.00), indicating a few higher-profit outliers. It means most profits are lower, with a few high-profit values affecting the average.
